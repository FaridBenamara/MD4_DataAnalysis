```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Charger le fichier CSV
file_path = "Stop_Data_2019_to_2022.csv"
df = pd.read_csv(file_path, low_memory=False)
```

```python
# Vérifier les valeurs manquantes en pourcentage
missing_values = df.isnull().sum() / len(df) * 100
missing_values = missing_values[missing_values > 0].sort_values(ascending=False)
print("\nValeurs manquantes par colonne:")
print(missing_values)
```

```python
# Supprimer les colonnes avec 100% de valeurs manquantes
df_cleaned = df.drop(columns=missing_values[missing_values == 100].index)
```

```python
# vérifier si des colonnes qui ont quasiment le meme nom sont les memes
 
df["PROP_SEARCH_OBJ_WARRANT"].equals(df["PROP_SRCH_OBJ_WARRANT"])
```

```python
df_diff = df[df["PROP_SEARCH_OBJ_WARRANT"] != df["PROP_SRCH_OBJ_WARRANT"]]
print(df_diff[["PROP_SEARCH_OBJ_WARRANT", "PROP_SRCH_OBJ_WARRANT"]].dropna())
```

```python
# Identifier et exclure les colonnes non numériques pour certaines analyses
non_numeric_cols = df_cleaned.select_dtypes(exclude=["number"]).columns
print("Colonnes non numériques:", non_numeric_cols)

df_numeric = df_cleaned.select_dtypes(include=["number"])
```

```python
# 1️⃣ Identifier les types d’arrêts
plt.figure()
df_cleaned["STOP_TYPE"].value_counts().plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Distribution des Types d'Arrêts")
plt.xlabel("Type d'Arrêt")
plt.ylabel("Nombre d'Arrêts")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2️⃣ Durée moyenne des arrêts
print("Durée moyenne des arrêts:", df_cleaned["STOP_DURATION_MINS"].mean())
print("Durée médiane des arrêts:", df_cleaned["STOP_DURATION_MINS"].median())
```

```python

# 3️⃣ Durée des arrêts par type
plt.figure()
df_cleaned.groupby("STOP_TYPE")["STOP_DURATION_MINS"].mean().plot(kind="bar", color="red")
plt.title("Durée Moyenne des Arrêts par Type")
plt.xlabel("Type d'Arrêt")
plt.ylabel("Durée Moyenne (minutes)")
plt.xticks(rotation=45)
plt.show()
```

```python
# 4️⃣ Principales raisons d’arrêt
plt.figure()
df_cleaned["STOP_REASON_NONTICKET"].value_counts().nlargest(10).plot(kind="bar", color="purple")
plt.title("Top 10 des Raisons d'Arrêt (Non-Ticket)")
plt.xlabel("Raison d'Arrêt")
plt.ylabel("Nombre d'Arrêts")
plt.xticks(rotation=90)
plt.show()
```

```python
# 5️⃣ Fréquence des fouilles
plt.figure()
df_cleaned[["PERSON_SEARCH_PROB_CAUSE", "PERSON_SEARCH_CONSENT"]].sum().plot(kind="bar", color=["blue", "orange"], edgecolor="black")
plt.title("Fréquence des Fouilles - Cause Probable vs Consentement")
plt.ylabel("Nombre Total")
plt.xticks(rotation=0)
plt.show()
```

```python
# 6️⃣ Répartition par genre
plt.figure()
df_cleaned["GENDER"].value_counts().plot(kind="bar", color="green", edgecolor="black")
plt.title("Répartition des Arrêts par Genre")
plt.xlabel("Genre")
plt.ylabel("Nombre d'Arrêts")
plt.show()
```

```python
# 7️⃣ Répartition par ethnie
plt.figure()
df_cleaned["ETHNICITY"].value_counts().plot(kind="bar", color="brown", edgecolor="black")
plt.title("Répartition des Arrêts par Ethnicité")
plt.xlabel("Ethnicité")
plt.ylabel("Nombre d'Arrêts")
plt.xticks(rotation=90)
plt.show()
```

```python

# 8️⃣ Contraventions vs Avertissements
plt.figure()
df_cleaned[["TICKET_COUNT", "WARNING_COUNT", "VOID_COUNT"]].sum().plot(kind="bar", color=["blue", "orange", "red"], edgecolor="black")
plt.title("Distribution des Contraventions, Avertissements et Annulations")
plt.ylabel("Nombre Total")
plt.xticks(rotation=0)
plt.show()
```

```python
# 9️⃣ Répartition géographique des arrêts
plt.figure()
df_cleaned["STOP_DISTRICT"].value_counts().plot(kind="bar", color="cyan", edgecolor="black")
plt.title("Répartition des Arrêts par District")
plt.xlabel("District")
plt.ylabel("Nombre d'Arrêts")
plt.xticks(rotation=45)
plt.show()
```

```python
# 🔟 Détection des valeurs aberrantes dans STOP_DURATION_MINS
plt.figure()
df_cleaned.boxplot(column=["STOP_DURATION_MINS"], vert=False)
plt.title("Boxplot de la Durée des Arrêts")
plt.xlabel("Durée (minutes)")
plt.show()
```

```python
# 🔥 Vérification des anomalies dans AGE et DATETIME
print("Valeurs uniques d'AGE suspectes:", df_cleaned["AGE"].unique())
print("Aperçu des dates d'arrêt suspectes:", df_cleaned["DATETIME"].head())
```

```python
# 🔥 Matrice de corrélation des colonnes numériques (originales)
plt.figure(figsize=(10,8))
cmap = plt.cm.coolwarm
plt.imshow(df_numeric.corr(), cmap=cmap, interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(df_numeric.columns)), df_numeric.columns, rotation=90)
plt.yticks(range(len(df_numeric.columns)), df_numeric.columns)
plt.title("Matrice de Corrélation des Variables Numériques (Originales)")
plt.show()
```
